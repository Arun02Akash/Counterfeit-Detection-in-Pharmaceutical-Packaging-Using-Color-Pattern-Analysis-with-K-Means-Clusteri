```python
from IPython.display import Image
imgpath = 'image.jpg'
Image(imgpath)
import numpy as np
from PIL import Image
img = Image.open('image.jpg')
img = img.resize((100,100))
img
img = np.array(img.convert('RGB'))
img = img.reshape((img.shape[0] * img.shape[1], 3))
img
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
scores = []
models = {}
for n in range(2,7):
 kmeans = KMeans(n_clusters = n)
 kmeans.fit(img)
 scores.append(silhouette_score(img, kmeans.labels_, metric='euclidean'))
 models[n]=kmeans
import matplotlib.pyplot as plt
plt.plot(list(range(2,7)),scores)
kmeans=models[4]
kmeans.cluster_centers_
(labels, freqs) = np.unique(kmeans.labels_, return_counts=True)
freqs = freqs/sum(freqs)
print(labels,freqs)
from PIL import Image
for label in labels:
 color = kmeans.cluster_centers_[label]
 color = tuple(int(x) for x in color)
 img = Image.new('RGB', (int(500*freqs[label]),50), color = color)
 display(img)
```
